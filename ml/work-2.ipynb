```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi*t)
```

```python
t = np.arange(0, 5, 0.1)

plt.subplot(211)

plt.subplot(t, f(t), 'bo')
plt.title("Régime Oscillatoires Amorties")

plt.subplot(212)
plt.plot(t, np.cos(2*np.pi*t), "r--");

plt.tight_layout()
```

# Histogramme

```python
mean  = 100
sigma = 15
# equivalent of code : mean + sigma * np.random.randn(10000)
x = np.random.normal(mean, sigma, 10000)

plt.hist(x, 50)

plt.xlabel('Valeur du QI')
plt.ylabel('Population')
plt.title('Répartition du Qi')
plt.axis([30, 180, 0, 800])
plt.grid()

plt.text(110, 600, r'$\mu=100,\ \sigma=15$', fontsize=8, bbox=dict(facecolor="cyan", alpha=0.2));
```

```python
text_file = path.expanduser('~/Documents/Ia-Datasets/moby.txt')
lettres = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

lcount = dict([(l,0) for l in lettres])

for l in open(text_file).read():
    try:
        lcount[l.upper()] += 1
    except KeyError:
        pass
    
print(lcount)
```

```python
total = sum(lcount.values())

print(total)
```

```python
fig = plt.figure(figsize=(8,4))

ax = fig.add_subplot(111)

ax.bar(range(len(lcount)), [lcount[l]/total * 100 for l in lettres], align='center')

ax.set_xticks(range(len(lcount)))
ax.set_xticklabels(lettres);
ax.set_ylabel('Freq %')
ax.set_title('Moby analysis')
```

```python

text_file2=path.expanduser('~/Documents/Ia-Datasets/hugo.txt')
lcount2=dict([(l,0) for l in lettres])
for l in open(text_file2).read():
    try:
        lcount2[l.upper()]+=1
    except KeyError:
        pass
total_lettres2=sum(lcount2.values())
bx=fig.add_subplot(212)
bx.bar(x,[lcount2[l]/total_lettres2*100 for l in lettres])
bx.set_xticks(x)
bx.set_xticklabels(lettres)
bx.set_ylabel('Fréquence de la lettre en %')
bx.set_title('Analyse de fréquence des Misérables en français')
plt.tight_layout();
```

## Etude d'une série temporelle : fremont

```python
data = pd.read_csv("~/Documents/Ia-Datasets/fremont.csv", index_col='Date', parse_dates=True)
data.head(5)
```

```python
data.info();
```

```python
data.tail(5) # 5 is defautl LOLL
```

```python
data.columns = ['East', 'West']

data.head()

```

```python
data['Total'] = data['East'] + data['West']

data.head()
```

```python
data.dropna(inplace=True)

data.isnull().sum()

data.info();

data.head();
```

```python
data.describe()
```

```python
sns.set()
```

```python
data.plot()
plt.ylabel('Comptage pietons')
```

```python
week = data.resample("W").sum()

week.plot()
plt.ylabel('Comptage pietons (base semaine)')
```

```python
day = data.resample('D').sum()
day.plot();
```

```python
day.rolling(60, center=True).sum().plot()
day.plot();
```

```python
per_hour = data.groupby(data.index.time).mean()

per_hour.plot()
```
