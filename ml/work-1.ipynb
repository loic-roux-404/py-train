```python
users.groupby('occupation').age.mean()
```

```python
users.groupby(['occupation', 'gender']).age.mean()
```

```python
users['genre_numerique'] = users.gender.apply(lambda x: 1 if x == 'M' else 0)
```

```python
# sum only take > 0 values so womans are excluded from subject data
percent = users.groupby('occupation').genre_numerique.sum() / users.occupation.value_counts() * 100

percent
```

```python
percent.sort_values(ascending=False)
```

```python
round(users.groupby(['gender', 'occupation']).age.agg(['min', 'mean', 'max']))
```
