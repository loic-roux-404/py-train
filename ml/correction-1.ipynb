```python
def transformation(x):
    if x=='M':
        return 1
    if x=='F':
        return 0
```

```python
users['genre_numerique']=users.gender.apply(transformation)
```

```python
users.head()
```

```python
users_pourcentage=users.groupby('occupation').genre_numerique.sum()/users.occupation.value_counts()*100
```

```python
users_pourcentage.sort_values(ascending=False) # % d'hommes par occupation
```
