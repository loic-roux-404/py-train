## Load the VGG16 Pre-trained Model

2. Faites des recherches sur ce modèle VGG16. Depuis quand date ce modèle ? Quelle est son architecture ? Sur quelles données se basent ce modèle ? etc

Ce modèle date de 2014 et à été le premier au classement de la compétition ILSVRC dans la catégorie "classification task".
Son architecture est faite d'un réseau de plusieurs couches de neurones convolutionnels de profondeurs accrues pour améliorer son efficacité avec entre chacun des filtres de poolings (5). Au total on a une pile de 3 couches connectés incluant chacun des chaines comme 4096 et 1000 et c'est seulement la dernière couche qui est une soft-max layer.

Le modèle se base sur des images de 1000 classes différentes triéesen trois jeux de 1.3 million images d'entrainement, 100,000 images de tests and 50,000 images de validation.

```python
# example of loading the vgg16 model
from keras.applications.vgg16 import VGG16
# load model
vgg16_model = VGG16()
# summarize the model
vgg16_model.summary()
```

# Prédiction

Vous allez tester le code ci-dessous avec chacune des photos (cat, dog et mug) et les deux modèles (VGG16 et ResNet50 et les prédictions associées).

```python
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array

images = []
for img in ["dog", "cat", "mug"]:
    # load an image from file
    image = load_img("%s.jpg" % img, target_size=(224, 224))
    # convert the image pixels to a numpy array
    image = img_to_array(image)
    # reshape data for the model
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    print(image)

    images.append(image)
```

6. Comment a été redimensionné l'image d'origine ?

Elle à été redimensionné grâce au reshape de keras pour avoir une taille conforme à ce que le modèle attend

```python
print("vgg16\n")
resnet_result = []
for image in images:
    # prepare the image for the VGG model
    image = preprocess_input(image)
    # predict the probability across all output classes
    yhat = vgg16_model.predict(image)
    # convert the probabilities to class labels
    label = decode_predictions(yhat)
    # retrieve the most likely result, e.g. highest probability
    label = label[0][0]
    resnet_result.append(label)
```

```python
# Resnet init
from keras.applications.resnet import ResNet50, preprocess_input, decode_predictions
```

```python
print('Resnet\n')
print("vgg16\n")
vgg16_result = []
for image in images:
    # prepare the image for the VGG model
    image = preprocess_input(image)
    # predict the probability across all output classes
    yhat = resnet_model.predict(image)
    # convert the probabilities to class labels
    label = decode_predictions(yhat)
    # retrieve the most likely result, e.g. highest probability
    label = label[0][0]
    vgg16_result.append(label)
```
