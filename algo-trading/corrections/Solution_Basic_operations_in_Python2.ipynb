```python
# Exercise 1: Basic File IO Operations
# Using Try - exception to load AAPL.csv and TSLA.csv
# implement calc_mean() and calc_var():
def calc_mean(seq):
    summation = 0
    for n in range(len(seq)):
        summation = summation + seq[n]
    avg = summation/len(seq)
    return avg

def calc_var(seq):
    avg = calc_mean(seq)
    summation = 0
    for n in range(len(seq)):
        summation = summation + (seq[n]-avg)**2
    var = summation/len(seq)
    return var

try: 
    f = open('AAPL.csv', 'r')
    f_result = open('result_AAPL.txt', 'w')
    raw = f.readlines(f)
    f.close()
    opening_price = []
    closing_price = []
    for r in raw:
        _, o_p, c_p = r.strip().split(',')
        opening_price.append(float(o_p))
        closing_price.append(float(c_p))
    avg_open = calc_mean(opening_price)
    var_open = calc_var(opening_price)
    avg_close = calc_mean(closing_price)
    var_close = calc_var(closing_price)
        
    f_result.write('The mean of opening price is {}\n'.format(avg_open))
    f_result.write('The variance of opening price is {}\n'.format(var_open))
    f_result.write('The mean of closing price is {}\n'.format(avg_close)
    f_result.write('The variance of closing price is {}\n'.format(var_close))
    f_result.close()
except IOError:
    print('The files does not exist.')
except:
    print('Unknown error occurs.')
```
