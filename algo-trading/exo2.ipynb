```python
# from exo1 import read_file, write_to_file
from typing import List
from utils import read_file
# Required : python -m pip install pandas
import pandas as pd


def panda_csv_describe(file: str, keys: List[str], index_by: str):
    try:
        df_covid = pd.read_csv(file)
    except IOError:
        print('The file does not exist.')
    except:
        print('Unknown error occurs.')

    for key in keys:
        df = df_covid.groupby(index_by)[key].describe()
        print(df)

def postals():
    metro_data = read_file("postalcode.txt")
    res = []
    for d in metro_data:
        _, postal = d.replace("\n", "").split(",")
        res.append(postal)

    print(res)

def descriptive_stats():
    panda_csv_describe(
        "Bordeaux-m√©tropole.csv",
        ["surface_terrain", "valeur_fonciere"],
        'code_postal'
    )

def main():
    postals()
    descriptive_stats()

main()
```
